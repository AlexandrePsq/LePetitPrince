```python
import pandas as pd
import numpy as np
import os
```

```python
for path in ['pearson_data.csv', 'r2_data.csv', 'significant_pearson_data.csv', 'significant_r2_data.csv']:
    df = pd.read_csv(path)
    df_group = df.groupby('model_names')
    result = []
    names = []
    for i, group in df_group:
        names.append(i)
        del group['model_names']
        del group['subject']
        df = pd.DataFrame(data=[np.mean(group.values, axis=0)], columns=group.columns)
        result.append(df)
    result = pd.concat(result, axis=0)
    result['model_names'] = names
    result.to_csv('averaged_'+path)
```
